# Build an Agentic AI Pipeline

This notebook demonstrates a simple agentic workflow where an AI agent understands user intent, selects an appropriate tool, and returns structured responses.

### Features

- Calculator Tool
- Keyword Extraction Tool
- Sentiment Analysis Tool
- Conditional Routing
- JSON Responses
- Error Handling

The project illustrates how a single-agent system can perform multiple responsibilities through tool selection and routing.

```python
def calculator(expression):

    try:
        return str(eval(expression))

    except:
        return "Calculation Error"
```

## Calculator Tool

The calculator tool processes mathematical expressions and returns the computed result.

```python
print(calculator("20 + 5"))
```

## Keyword Extraction Tool

This tool extracts important keywords from a sentence by filtering meaningful words.

```python
def extract_keywords(text):

    words = text.split()

    keywords = list(
        set(
            [
                word.lower().strip(".,!?")
                for word in words
                if len(word) > 4
            ]
        )
    )

    return keywords[:5]
```

## Testing Keyword Extraction

The following example demonstrates how keywords are extracted from text.

```python
sample_text = "Artificial Intelligence is transforming industries"

print(extract_keywords(sample_text))
```

## Sentiment Analysis Tool

This tool identifies whether a sentence expresses positive, negative, or neutral sentiment.

```python
def sentiment_analysis(text):

    positive_words = [
        "good",
        "great",
        "excellent",
        "happy",
        "awesome"
    ]

    negative_words = [
        "bad",
        "poor",
        "sad",
        "terrible",
        "worst"
    ]

    text = text.lower()

    if any(word in text for word in positive_words):
        return "Positive"

    if any(word in text for word in negative_words):
        return "Negative"

    return "Neutral"
```

## Testing Sentiment Analysis

The tool classifies user text into positive, negative, or neutral categories.

```python
print(sentiment_analysis("This project is awesome"))
```

## Agent Routing Logic

The agent analyzes the query and routes it to the appropriate tool based on user intent.

Routing Rules:

- Calculate → Calculator Tool
- Keyword → Keyword Extraction Tool
- Sentiment → Sentiment Analysis Tool
- Other queries → General Response

```python
def agent(query):

    try:

        query_lower = query.lower()

        if "calculate" in query_lower:

            expression = query_lower.replace(
                "calculate",
                ""
            ).strip()

            return {
                "type": "calculation",
                "result": calculator(expression)
            }

        elif "keyword" in query_lower:

            text = (
                query
                .replace("Extract keywords from", "")
                .replace("extract keywords from", "")
                .strip()
            )

            return {
                "type": "keywords",
                "result": extract_keywords(text)
            }

        elif "sentiment" in query_lower:

            text = query.replace(
                "sentiment",
                ""
            ).strip()

            return {
                "type": "sentiment",
                "result": sentiment_analysis(text)
            }

        else:

            return {
                "type": "general",
                "result": f"Received query: {query}"
            }

    except Exception as e:

        return {
            "type": "error",
            "result": str(e)
        }
```

## Agent Testing

The following test cases verify whether the agent correctly identifies user intent and selects the appropriate tool.

```python
queries = [
    "Calculate 25 + 10",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "Sentiment This project is awesome",
    "What is machine learning?"
]

for query in queries:

    print("Query:", query)

    print(agent(query))

    print("-" * 50)
```

## Error Handling

Error handling improves reliability by preventing the system from failing when invalid inputs are provided.

```python
print(agent("Calculate 10 / 0"))
```

## Interactive Testing

Users can enter custom queries and observe how the agent dynamically selects tools and generates responses.

```python
sample_queries = [
    "Calculate 15 + 25",
    "Extract keywords from Data Science and Artificial Intelligence are popular fields",
    "Sentiment This project is excellent"
]

for query in sample_queries:

    print("Query:", query)

    print("Response:", agent(query))

    print("-" * 50)
```
